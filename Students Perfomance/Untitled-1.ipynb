```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
data= pd.read_csv("StudentsPerformance.csv")
print(data)
```

Data Claening processe

```python
print(data.columns)
```

```python
print(data.isnull().sum())
```

```python
Ndata= data.drop(["test preparation course", "race/ethnicity"], axis=1)
print(Ndata)
```

Data Preprocessing

```python
object_data= Ndata.select_dtypes(include=["object"])
print(object_data)
le= preprocessing.LabelEncoder()
for i in range(object_data.shape[1]):
    object_data.iloc[:,i]= le.fit_transform(object_data.iloc[:,i])

print(object_data)    
num_data= Ndata.select_dtypes(exclude=["object"])
Ndata= pd.concat([object_data,num_data], axis=1)
```

Corr

```python
x= Ndata.corr()
print (x)
sns.heatmap(x, annot=True)
plt.show()
```
